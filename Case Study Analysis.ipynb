## Data Discovery and Visualization

### How does the number of claims differ across locations?

We have a total of 98 cases across 4 locations where claims are processed. Sao Luis has the highest total of 30 cases. Recife has the second-highest total of 25, followed by Fortaleza with 22 and Natal with 21 cases. 

![NameofImage](TotalClaimsbyLocation.png)         

### What is the distribution of time to close claims?

As the legal team spans multiple locations, we should look at the distribution of time to close cases over the 10-year period. As we can see in the following chart, the total time to close claims has decreased significantly from 2010-2020. As a reference, in 2010 the longest case took 3591 days (9 years, 8 months). In 2020, the longest case took 349 days (just under one year). 

![NameofImage](Dist.png)   

Across the 10-year period, only 8 cases took longer than 2000 days to close (5 years and 4 months). This was only seen in cases prior to 2016. 

![NameofImage](IQR.png)   

### How does the average time to close claims differ by location?

Sao Luis has the highest average time to close (1030 days/2.8 years), followed by Recife (861 days/2.3 years), Fortaleza (776/2.1 years), and Natal (581 days/1.5 years). We should also consider that each location's performance may vary on a year-by-year basis. A valuable consideration when reviewing this data is that during the first 4 years, the workload was not distributed across all 4 locations. From 2010 - 2013, Sao Luis, Recife and Fortaleza were the only locations processing claims. It was only in 2016 that all 4 locations began processing cases.

![NameofImage](avgttc.png)   

The chart below represents a breakdown of the time to close claims, by location from 2016 - 2020. 2016 was the first year to see all 4 locations processing claims. Three locations have seen a year-on-year improvement in the average time to close each case assigned to them with the exception of Natal. The head of legal may want to investigate the possible cause of Natal's superior performance in 2016. This is followed by a dramatic increase in the average time to settle claims in 2017. Natal has however since recovered and joined the trend of year-on-year performance improvement from 2018 - 2020. 

![NameofImage](year-location.png)   

Although there has been an improvement in the time to close each case, it is unfortunate that of the 98 cases presented in this analysis, 78 of the claims have the cause of food poisoning listed as ‘unknown’. A more accurate recording of causes would be valuable in conducting further analysis on whether the cause of claims has any impact on the time to close.

In the final chart, we can see that claim settlement costs have shot up significantly since 2016. The legal team should encourage further investigations to determine the underlying cause of each case. Addressing these causes would lead to a reduction in claims and a reduction in costs.
